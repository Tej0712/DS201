```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Lemonade2016-2.csv')
```

```python
df
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.reset_index(drop=True)
df
```

```python
df = df.drop(7)
df=df.reset_index(drop=True)
df
```

```python
df['Leaflets'].mean()
```

```python
avg_value=round(df['Leaflets'].mean(),0)
df['Leaflets']=df['Leaflets'].fillna(avg_value)
df
```

```python
df['Sales']=df['Lemon']+df['Orange']
df['Revenue'] = df['Sales']*df['Price']
df
```

```python
df.plot.scatter(x="Leaflets",y="Sales")
```

```python
df["Revenue"].plot.hist(bins=10,title='Revenue Histogram of Lemonade Sales')
```

```python
df.boxplot(by='Location',column='Sales')
```

```python
ax=sns.boxplot(x="Location",y="Sales",data=df,)
```

```python
ax=sns.boxplot(x="Location",y="Sales",hue="Location", data=df,palette="Set3")
```

```python
df1 = df.loc[df['Temperature']>=75,:]
df1
```

```python
ax=sns.boxplot(x="Location",y="Sales",hue="Location",
               data=df1,palette="Set3")
```
