```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/asukul/DS201/master/datasets/Lemonade2016-2.csv")
```

```python
df
```

```python
df=df.drop_duplicates(ignore_index=True)
```

```python
df
```

```python
avg_value=round(df['Leaflets'].mean(),0)
```

```python
df['Leaflets']=df['Leaflets'].fillna(avg_value)
df
```

```python
df['Sales']=df['Lemon']+df['Orange']
df['Revenue'] = df['Sales']*df['Price']
df
```

```python
df.reset_index(drop=True)
```

```python
from bokeh.plotting import figure,show, output_file
p=figure(title="Sales by temperature with revenue as circle size",x_axis_label="Temperature",y_axis_label="Sales")
output_file("Poola-DS201X-HW5-plot1.html")
p.circle(df["Temperature"],df["Sales"],legend_label="Sales",fill_color="Yellow",line_color="Black",size=df["Revenue"]/2,fill_alpha=0.5)
p.legend.location="top_left"
show(p)
```

```python
from bokeh.plotting import figure, show, output_file, gridplot

p1 = figure(title="Revenue by Leaflets", x_axis_label="Leaflets", y_axis_label="Revenue")
p1.scatter(df["Leaflets"], df["Revenue"], size=10, color="blue")

p2 = figure(title="Revenue vs. Sales", x_axis_label="Sales", y_axis_label="Revenue")
p2.scatter(df["Sales"], df["Revenue"], size=10, color="green")

p3 = figure(title="Sales by Temperature", x_axis_label="Temperature", y_axis_label="Sales")
p3.scatter(df["Temperature"], df["Sales"],size=10, color="red")

grid = gridplot([[p1, p2, p3]])


output_file("Poola-DS201X-HW5-plot2.html",mode="inline")

show(grid)
```
