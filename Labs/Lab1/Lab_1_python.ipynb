```python
import matplotlib.pyplot as plt
import pandas as pd
low_memory = False
```

```python
#Load dataset
```

```python
file_data = pd.read_csv('Most-Recent-Cohorts-Scorecard-Elements.csv')
```

```python
#Print head lines of dataset
```

```python
file_data.head(10)
```

```python
#Group dataset
```

```python
file = file_data.groupby(['STABBR']).size()
```

```python
file.head()
```

```python
#Plot chart of database with panda.DataFrame.plot.bar
```

```python
ax = file.plot.bar(figsize=[10,8], color='skyblue', edgecolor='black')
ax.set_title('Number of schools per state')
ax.set_xlabel('State')
ax.set_ylabel('Number of Schools')
```

```python
#Plot chart of dataset with pandas.DataForm.plot.bar and make red colomn 'IA'
```

```python
ax = file.plot.bar(figsize=[10,8], color='skyblue', edgecolor='black')
ax.set_title('Number of schools per state')
ax.set_xlabel('State')
ax.set_ylabel('Number of Schools')

ia_color = 'red'
ia_index = file.index.get_loc('IA')
ax.patches[ia_index].set_facecolor(ia_color)

plt.show()
```
