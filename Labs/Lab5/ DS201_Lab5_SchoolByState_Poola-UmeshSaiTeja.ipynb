```python
df2 = df.groupby(['State_name']).size()
df2
```

```python
df2.plot.bar()
plt.show()
```

```python
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'orange', 'purple', 'brown', 'pink']

df2.plot.bar(figsize=(15,7), color=colors)

plt.title("Number of schools per states", fontsize =15)
plt.xlabel("States", fontsize = 15)
plt.ylabel("Number of schools", fontsize = 15)
plt.text(4,800, "Maximum", fontsize = 15)
plt.show()
```
